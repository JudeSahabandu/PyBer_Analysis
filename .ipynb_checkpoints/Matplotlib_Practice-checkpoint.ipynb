```python
# To make sure our plots are displayed inline within the Jupyter Notebook directly below the code cell that produced it
# % is the magic command and inline ensures plots are in the output cell
%matplotlib inline
```

```python
# Import dependencies
# The pyplot module will allow us to create all the necessary charts
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month. Use arrays[]
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month. Use arrays[]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plot()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis,y_axis)
```

```python
# To change figure-level attributes (such as axis labels, a title, or a legend) or save the figure as an image
fig = plt.figure()
```

```python
#Setting figure size
plt.figure(figsize = (7,5))

# Create a label, marker, color and linewidth for the legend
plt.plot(x_axis, y_axis, label = "Boston", marker = "o", color = "blue", linewidth = 2.5)

# Create a legend and a grid
plt.legend()
plt.grid()

# Create x axis label, y axis label and title
plt.xlabel("Date")
plt.ylabel("Fare ($)")
plt.title("Pyber Fare by Month")

# Set the y limit between 0 and 45.
plt.ylim(0,45)
```

```python
#Setting figure size
fig, ax = plt.subplots(figsize = (7,5))

# Create a label, marker, color and linewidth for the legend
ax.plot(x_axis, y_axis, label = "Boston", marker = "D", color = "green", linewidth = 2)

# Create a legend and a grid
ax.legend()
ax.grid()

# Create x axis label, y axis label and title
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.set_title("Pyber Fare by Month")

# Set the y limit between 0 and 45.
ax.set_ylim(0,45)
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot
plt.bar(x_axis, y_axis, color = "green", label = "Boston")

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Create a title.
plt.title("PyBer Fare by Month")

# Add a legend
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis, color = "magenta", label = "Boston")

# Invert values of y axis
plt.gca().invert_yaxis()

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Create a title.
plt.title("PyBer Fare by Month")
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis, color = "cyan", label = "Chicago")
ax.legend()

ax.set_xlabel("Fare ($)")
ax.set_ylabel("Month")
ax.set_title("Pyber Fare by Month")

plt.gca().invert_yaxis()
```

```python
plt.plot(x_axis, y_axis, "o")
```

```python
plt.scatter(y_axis, x_axis, c="red", label = "Chicago")
plt.legend()

plt.xlabel("Fare ($)")
plt.ylabel("Month")
plt.title("Pyber Fare by Month")

plt.xlim(0,50)
plt.gca().invert_yaxis()
```

```python
# Setting scatter size using for loops

y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s=y_axis_larger)
plt.show()
```

```python
# Python's list comprehension technique to replace many for and while loops
# new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()

ax.scatter(y_axis, x_axis, c = "skyblue", edgecolor = "black", linewidth = 2, s = [i * 5 for i in y_axis], label = "Boston", alpha = 0.75)

ax.legend()

ax.set_title("Pyber Fare By Month")
ax.set_xlabel("Months")
ax.set_ylabel("Fare ($)")

plt.gca().invert_yaxis()

plt.xlim(0,50)
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# Resize pie chart
plt.subplots(figsize=(10,10))

# Create list/tuple for exploding chart
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%', colors=colors)

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,8))

explode_list = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)

ax.pie(y_axis, labels = x_axis, autopct = '%.1f%%', explode = explode_list, shadow = 'True', startangle = 90, colors = colors)

plt.gca().invert_xaxis()

plt.show()
```
