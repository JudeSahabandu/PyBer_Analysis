# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_city_type

# Obtaining Line Graph for Total Ride Data by City Type

total_rides = total_rides_city_type.plot(figsize = (15,6))

total_rides.set_title("Total Rides by City Type")
total_rides.set_ylabel("Total Rides")

plt.tight_layout()
plt.savefig("Analysis/PyBer_total_rides.png")
```

```python
# 2. Get the total drivers for each city type
total_drivers_city_type = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_city_type

# Obtaining Line Graph for Total Driver Data by City Type

total_drivers = total_drivers_city_type.plot(figsize = (15,6))

total_drivers.set_title("Total Drivers by City Type")
total_drivers.set_ylabel("Total Drivers")

plt.tight_layout()
plt.savefig("Analysis/PyBer_total_drivers.png")
```

```python
#  3. Get the total amount of fares for each city type
total_fare_city_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fare_city_type

# Obtaining Line Graph for Total Fare Data by City Type

total_fare = total_fare_city_type.plot(figsize = (15,6))

total_fare.set_title("Total Fare by City Type")
total_fare.set_ylabel("Total Fare ($USD)")

plt.tight_layout()
plt.savefig("Analysis/PyBer_total_fare.png")
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_per_ride = (total_fare_city_type / total_rides_city_type)
average_fare_per_ride

# Obtaining Line Graph for Avg Fare per Ride Data by City Type

average_fare_ride = average_fare_per_ride.plot(figsize = (15,6))

average_fare_ride.set_title("Ride Avg. Fare by City Type")
average_fare_ride.set_ylabel("Avg. Fare ($USD)")

plt.tight_layout()
plt.savefig("Analysis/PyBer_avg_fare_ride.png")
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver = (total_fare_city_type / total_drivers_city_type)
average_fare_per_driver

# Obtaining Line Graph for Avg Fare per Driver Data by City Type

average_fare_driver = average_fare_per_driver.plot(figsize = (15,6))

average_fare_driver.set_title("Driver Avg. Fare by City Type")
average_fare_driver.set_ylabel("Avg. Fare ($USD)")

plt.tight_layout()
plt.savefig("Analysis/PyBer_avg_fare_driver.png")
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame ({"Total Rides":total_rides_city_type,
                                  "Total Drivers":total_drivers_city_type,
                                  "Total Fares":total_fare_city_type,
                                  "Average Fare Per Ride":average_fare_per_ride,
                                 "Average Fare Per Driver":average_fare_per_driver})

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare Per Ride"] = pyber_summary_df["Average Fare Per Ride"].map("${:.2f}".format)
pyber_summary_df["Average Fare Per Driver"] = pyber_summary_df["Average Fare Per Driver"].map("${:.2f}".format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fare_sum_city_date = pyber_data_df.groupby(["type", "date"]).sum()["fare"]
fare_sum_city_date
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
fare_sum_city_date = fare_sum_city_date.reset_index()
fare_sum_city_date
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pivot_fare_sum_city_date = fare_sum_city_date.pivot(index='date', columns='type', values='fare')
pivot_fare_sum_city_date
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
citytype_fare_by_date = pivot_fare_sum_city_date.loc['2019-01-01': '2019-04-29']
citytype_fare_by_date
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
citytype_fare_by_date.index = pd.to_datetime(citytype_fare_by_date.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
citytype_fare_by_date.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fare_by_citytype = citytype_fare_by_date.resample('W').sum()
weekly_fare_by_citytype
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

weekly_fares = weekly_fare_by_citytype.plot(figsize = (15,6))

# Line graph title and y-axis label

weekly_fares.set_title("Total Fare by City Type")
weekly_fares.set_ylabel("Fare ($USD)")


# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

plt.tight_layout()

# Saving Image
plt.savefig("Analysis/PyBer_fare_summary.png")

plt.show()

```
